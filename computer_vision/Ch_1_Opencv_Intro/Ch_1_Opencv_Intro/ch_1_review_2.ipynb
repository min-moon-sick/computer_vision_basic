```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import cv2

img_list = os.listdir('./fig/my_images/')

imgs = []

for i in img_list:
    imgs.append('./fig/my_images/' + i)

print(imgs)
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0
while 1:
    img = cv2.imread(imgs[idx])
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(2000) == 27:
        break
    
    if idx == len(imgs):
        idx = 0
        
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import cv2

src = cv2.imread('/Users/HPE/Documents/computer_vision/Ch_2_Opencv_basic/Ch_2_Opencv_basic/fig/puppy.bmp')
sunglass = cv2.imread('/Users/HPE/Documents/computer_vision/Ch_2_Opencv_basic/Ch_2_Opencv_basic/fig/imgbin_sunglasses_1.png', cv2.IMREAD_UNCHANGED)

sunglass = cv2.resize(sunglass, (300, 150))

mask = sunglass[: ,:, -1]
glass = sunglass[: ,:, :-1]

h, w = mask.shape[:2]
src_crop = src[120: 120 + h, 220: 220 + w]

cv2.copyTo(glass, mask, src_crop)

cv2.imshow('src', src)
cv2.imshow('sunglass', sunglass)
cv2.waitKey()
cv2.destroyAllWindows()
```
