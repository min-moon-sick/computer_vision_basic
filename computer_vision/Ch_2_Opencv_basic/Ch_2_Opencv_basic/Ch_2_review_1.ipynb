```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import cv2

src = cv2.imread('./fig/cow.png')
dst = cv2.imread('./fig/green.png')

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

ret, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)
# print(ret)

h, w = mask.shape[:2]

crop = dst[300: 300 + h, 400: 400 + w]

cv2.copyTo(src, mask, crop)

cv2.namedWindow('src')
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('src_gray', src_gray)
cv2.imshow('mask', mask)

while 1:
    key = cv2.waitKey()
    
    if key == 27:
        break
        
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import cv2

src = cv2.imread('./fig/imgbin_sunglasses_1.png', cv2.IMREAD_UNCHANGED)
dst = cv2.imread('./fig/cow.png')
print(src.shape)

h, w = src.shape[:2]
src = cv2.resize(src, (w // 8, h // 8))

mask = src[:, :, -1]
img  = src[:, :, :-1]

h, w = mask.shape[:2]

dst_crop = dst[100: 100 + h, 200: 200 + w]

cv2.copyTo(img, mask, dst_crop)
cv2.namedWindow('src')

cv2.imshow('src', src)
cv2.imshow('mask', mask)
cv2.imshow('dst', dst)
cv2.imshow('dst_crop', dst_crop)

while 1:
    key = cv2.waitKey()
    
    if key == 27:
        break
        
cv2.destroyAllWindows()
```

```python
## mouse callback 
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import cv2

oldx = oldy = 0

def call_mouse(event, x, y, flags, param):
    global oldx, oldy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        print(x, y)
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y),
                    (0, 0, 255), 3, cv2.LINE_AA)
            cv2.imshow('img', img)
            oldx, oldy = x, y
        
        
        
img = np.ones((480 , 640, 3), np.uint8) * 255

cv2.namedWindow('img')

cv2.imshow('img', img)
cv2.setMouseCallback('img', call_mouse, img)

while 1:
    key = cv2.waitKey()
    
    if key == 27:
        break
        
cv2.destroyAllWindows()
```

```python
## trackbar

import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import cv2

oldx = oldy = 0

def call_track(pos):
#     print(pos)
    img[:] = pos
    cv2.imshow('img', img)
        
        
        
img = np.zeros((480 , 800), np.uint8)

cv2.namedWindow('img')

cv2.imshow('img', img)
cv2.createTrackbar('level', 'img', 50, 255, call_track)
cv2.setMouseCallback('img', call_mouse, img)

while 1:
    key = cv2.waitKey()
    
    if key == 27:
        break
        
cv2.destroyAllWindows()
```
